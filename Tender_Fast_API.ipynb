```python
# model_manager.py

import asyncio
import torch
from huggingface_hub import login
from llama_index.embeddings.fastembed import FastEmbedEmbedding
from llama_index.core import VectorStoreIndex, Settings
from llama_index.llms.huggingface import HuggingFaceLLM
from llama_index.core.postprocessor import (
    SentenceTransformerRerank,
    KeywordNodePostprocessor,
)
from llama_index.vector_stores.qdrant import QdrantVectorStore
from qdrant_client import QdrantClient
from transformers import AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig
import os

class ModelManager:
    _instance = None  # Singleton instance

    def __new__(cls):
        if cls._instance is None:
            cls._instance = super(ModelManager, cls).__new__(cls)
            # Initialize variables
            cls._instance.llm = None
            cls._instance.embed_model = None
            cls._instance.rerank = None
            cls._instance.query_engines = {}  # Map from user_id to query_engine
            cls._instance.lock = asyncio.Lock()
            cls._instance.inference_lock = asyncio.Lock()  # Added inference lock for thread safety
        return cls._instance

    def load_models(self):
        try:
            # Hugging Face login
            # Use environment variables or secure storage for tokens
            hf_token = "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"  # Replace with your actual token or use environment variables
            login(token=hf_token)
            print("Logged in successfully!")

            quantization_config = BitsAndBytesConfig(
                load_in_4bit=True,
                bnb_4bit_quant_type="nf4",
                bnb_4bit_compute_dtype=torch.bfloat16
            )

            NEW_MODEL = "Arun-PEMS/GEM_Train_1"
            tokenizer = AutoTokenizer.from_pretrained(NEW_MODEL, use_fast=True)

            model = AutoModelForCausalLM.from_pretrained(
                NEW_MODEL,
                quantization_config=quantization_config,
                device_map="auto"
            )

            # Initialize the HuggingFace LLM
            self.llm = HuggingFaceLLM(
                model_name=NEW_MODEL,
                tokenizer=tokenizer,
                model=model,
            )

            # Set the LLM and embed_model in Settings
            self.embed_model = FastEmbedEmbedding(model_name="BAAI/bge-small-en-v1.5")
            Settings.llm = self.llm
            Settings.embed_model = self.embed_model

            # Define the rerank postprocessor
            self.rerank = SentenceTransformerRerank(
                model="cross-encoder/ms-marco-MiniLM-L-2-v2",
                top_n=3
            )

            print("Models loaded successfully")
        except Exception as e:
            print(f"Error during model loading: {e}")
            raise

    async def create_query_engine(self, user_id, required_keywords):
        if self.llm is None or self.embed_model is None or self.rerank is None:
            raise Exception("Models are not loaded. Please load models first.")

        async with self.lock:
            # Check if query engine already exists
            if user_id in self.query_engines:
                return

            # Create query engine asynchronously
            await self._create_query_engine_async(user_id, required_keywords)

    async def _create_query_engine_async(self, user_id, required_keywords):
        try:
            # Qdrant credentials
            # Use environment variables or secure storage for API keys

            client = QdrantClient(
            url="https://fd4ba2d3-e45f-44d6-9b20-b0c5d1cabb9d.us-east4-0.gcp.cloud.qdrant.io:6333",
            api_key="y2EIzplsyAyYJPur5ryVUldfHtT6Q1-rM48c3uVLqsYBhmKk1qPhnQ",
        )


            # Create the vector store using the existing collection
            vector_store = QdrantVectorStore(
                client=client,
                collection_name="GEM_2_document",
            )

            # Load the index from the vector store
            index = VectorStoreIndex.from_vector_store(
                vector_store=vector_store
            )

            # Create the query engine with user-specific required keywords
            query_engine = index.as_query_engine(
                similarity_top_k=10,
                node_postprocessors=[
                    KeywordNodePostprocessor(required_keywords=[required_keywords]),
                    self.rerank
                ]
            )

            # Store the query engine for the user
            self.query_engines[user_id] = query_engine

            print(f"Query engine created successfully for user {user_id}")
        except Exception as e:
            print(f"Error during query engine creation: {e}")
            raise
```

```python
# app.py

from fastapi import FastAPI, HTTPException
from pydantic import BaseModel
import time
import asyncio
import os

from model_manager import ModelManager

app = FastAPI()

# Create a singleton instance of ModelManager
model_manager = ModelManager()

# Pydantic models for the requests
class QueryRequest(BaseModel):
    user_id: str
    query: str

class CreateQueryEngineRequest(BaseModel):
    user_id: str
    required_keywords: str

@app.post("/load_models")
async def load_models_endpoint():
    try:
        # Run the blocking load_models method in a separate thread to avoid blocking the event loop
        await asyncio.to_thread(model_manager.load_models)
        return {"message": "Models loaded successfully"}
    except Exception as e:
        raise HTTPException(status_code=500, detail=str(e))

@app.post("/create_query_engine")
async def create_query_engine_endpoint(request: CreateQueryEngineRequest):
    try:
        await model_manager.create_query_engine(
            user_id=request.user_id,
            required_keywords=request.required_keywords
        )
        return {"message": f"Query engine created successfully for user {request.user_id}"}
    except Exception as e:
        raise HTTPException(status_code=500, detail=str(e))

@app.get("/health_check")
async def health_check():
    try:
        if (model_manager.llm is not None and
            model_manager.rerank is not None and
            model_manager.query_engines):
            return {"status": "Models and query engines are loaded"}
        else:
            return {"status": "Models and query engines are not fully loaded"}
    except Exception as e:
        raise HTTPException(status_code=500, detail=str(e))

@app.post("/query")
async def run_query(request: QueryRequest):
    print(f"Request Received: {request.query}")
    try:
        query_engine = model_manager.query_engines.get(request.user_id)
        if query_engine is None:
            raise HTTPException(
                status_code=500,
                detail=f"Query engine is not initialized for user {request.user_id}"
            )

        now = time.time()

        # Acquire the inference lock to prevent concurrent access
        async with model_manager.inference_lock:
            # Run the synchronous query method in a separate thread
            response = await asyncio.to_thread(query_engine.query, request.query)

        elapsed_time = round(time.time() - now, 2)
        return {
            "response": str(response),
            "elapsed_time": f"{elapsed_time} seconds"
        }
    except Exception as e:
        raise HTTPException(status_code=500, detail=str(e))
```

```python
# Save app.py
with open('app.py', 'w') as f:
    f.write("""
from fastapi import FastAPI, HTTPException
from pydantic import BaseModel
import time
import asyncio
import httpx  # For HTTP requests

from model_manager import ModelManager

app = FastAPI()

# Create a singleton instance of ModelManager
model_manager = ModelManager()

# Pydantic models for the requests
class QueryRequest(BaseModel):
    user_id: str
    query: str

class CreateQueryEngineRequest(BaseModel):
    user_id: str
    required_keywords: str

@app.post("/load_models")
async def load_models_endpoint():
    try:
        # Run the blocking load_models method in a separate thread to avoid blocking the event loop
        await asyncio.to_thread(model_manager.load_models)
        return {"message": "Models loaded successfully"}
    except Exception as e:
        raise HTTPException(status_code=500, detail=str(e))

@app.post("/create_query_engine")
async def create_query_engine_endpoint(request: CreateQueryEngineRequest):
    try:
        await model_manager.create_query_engine(
            user_id=request.user_id,
            required_keywords=request.required_keywords
        )
        return {"message": f"Query engine created successfully for user {request.user_id}"}
    except Exception as e:
        raise HTTPException(status_code=500, detail=str(e))

@app.get("/health_check")
async def health_check():
    try:
        # Check if the models and query engines are loaded
        if model_manager.embed_model is not None and model_manager.query_engines:
            return {"status": "Models and query engines are loaded"}
        else:
            return {"status": "Models and query engines are not fully loaded"}
    except Exception as e:
        raise HTTPException(status_code=500, detail=str(e))

@app.post("/query")
async def run_query(request: QueryRequest):
    print(f"Request Received from {request.user_id}: {request.query}")
    try:
        query_engine = model_manager.query_engines.get(request.user_id)
        if query_engine is None:
            raise HTTPException(
                status_code=500,
                detail=f"Query engine is not initialized for user {request.user_id}"
            )

        now = time.time()

        # Prepare the payload for vLLM
        payload = {
            "prompt": request.query,
            "max_tokens": 256,  # Adjust as needed
            "temperature": 0.7,  # Adjust as needed
            "top_p": 0.95,       # Adjust as needed
            "n": 1               # Number of responses
        }

        # Send the request to the vLLM server
        async with httpx.AsyncClient() as client:
            vllm_response = await client.post("http://localhost:8000/generate", json=payload)
            vllm_response.raise_for_status()
            vllm_result = vllm_response.json()

        # Extract the generated text
        generated_text = vllm_result["outputs"][0]["text"]

        elapsed_time = round(time.time() - now, 2)
        return {
            "response": generated_text,
            "elapsed_time": f"{elapsed_time} seconds"
        }
    except Exception as e:
        raise HTTPException(status_code=500, detail=str(e))
""")
```
