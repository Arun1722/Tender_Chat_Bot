```python
import re
import unicodedata
from langdetect import detect

def is_english(text):
    try:
        return detect(text) == 'en'
    except:
        return False

def preprocess_text(text):
    # Normalize text to remove special unicode characters
    text = unicodedata.normalize('NFKD', text)

    # Regular expression to find dates in the format dd-mm-yyyy or yyyy-mm-dd
    date_pattern = re.compile(r'\b\d{2}-\d{2}-\d{4}\b|\b\d{4}-\d{2}-\d{2}\b')

    # Find all dates and replace them with a placeholder
    dates = date_pattern.findall(text)
    text = date_pattern.sub(' <DATE> ', text)

    # Remove special characters except dates, alphanumeric and whitespace
    text = re.sub(r'[^A-Za-z0-9\s<DATE>]', '', text)

    # Replace placeholder with the original dates
    for date in dates:
        text = text.replace('<DATE>', date, 1)

    # Replace multiple spaces with a single space and trim leading/trailing spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # Filter out non-English text
    if not is_english(text):
        return ''

    return text

def preprocess_elements(elements):
    processed_elements = []
    for element in elements:
        if element.type in ['text', 'table']:
            cleaned_text = preprocess_text(element.text)
            if cleaned_text:
                processed_elements.append({
                    'type': element.type,
                    'text': cleaned_text
                })
    return processed_elements


# Sample input
#text_elements = [Element(type='text', text='Example text with special characters !@#$%^&*() and newlines \n\n')]
#table_elements = [Element(type='table', text='Example table text with special characters !@#$%^&*() and newlines \n\n')]

# Preprocess elements
preprocessed_text_elements = preprocess_elements(text_elements)
preprocessed_table_elements = preprocess_elements(table_elements)

# Output results
print(preprocessed_text_elements[2])
print(preprocessed_table_elements[3])
```

```python
def get_vector(tir_client, text):
    """
    Get vector representation of text using the embedding model.

    Args:
        tir_client (tir.ModelAPIClient): The TIR client.
        text (str): The input text to be vectorized.

    Returns:
        list: The vector representation of the text.
    """
    data = {"prompt": text}
    response = tir_client.infer(model_name=EMBEDDING_MODEL_NAME, data=data)
    # Add debug print to inspect response structure
    print("TIR Client Response:", response)
    return response.outputs[0].data

def retrieve_context(vector):
    """
    Perform vector search in Qdrant and retrieve context.

    Args:
        vector (list): The query vector.

    Returns:
        str: The concatenated context from the search results.
    """
    search_result = qdrant_client.search(
        collection_name=QDRANT_COLLECTION_NAME,
        query_vector=vector,
        limit=3,  # Get top 3 vectors based on the search score
    )

    # Add debug print to inspect search_result structure
    print("Search result from Qdrant:", search_result)

    # Extract context from search results
    payloads = [hit.payload for hit in search_result]
    context = ' '.join(hit['data'] for hit in payloads)
    return context

# Define the prompt template
template = """Answer the question based only on the following context, which can include text and tables:
{context}
Question: {question}
"""
prompt_template = ChatPromptTemplate.from_template(template)

def chat_with_E2E(tir_client, question):
    """
    Handle the full chat interaction, from encoding the prompt to generating a response.

    Args:
        tir_client (tir.ModelAPIClient): The TIR client.
        question (str): The user's input question.
    """
    try:
        # Encode question to get its vector representation
        vector = get_vector(tir_client, question)

        # Retrieve context using the vector
        context = retrieve_context(vector)

        # Define the RAG pipeline chain
        chain = LLMChain(
            {"context": RunnablePassthrough(), "question": RunnablePassthrough()},
            prompt=prompt_template,
            llm=model,
            output_parser=StrOutputParser()
        )

        # Invoke the chain with the question and context
        response = chain.invoke({"context": context, "question": question})
        print(response)
    except Exception as e:
        print(f"An error occurred: {e}")

# Example usage
chat_with_E2E(tir_client, "What are the additional specification documents?")
```
