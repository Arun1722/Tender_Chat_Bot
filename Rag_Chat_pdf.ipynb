```python
%%writefile chat.py
import streamlit as st
from unstructured.partition.pdf import partition_pdf
from langchain.retrievers.multi_vector import MultiVectorRetriever
from langchain.storage import InMemoryStore
from langchain_community.vectorstores import Chroma
from langchain_core.documents import Document
from langchain_community.embeddings.fastembed import FastEmbedEmbeddings
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.runnables import RunnablePassthrough
from langchain_community.llms.huggingface_pipeline import HuggingFacePipeline
import torch
import uuid
import nltk
import tempfile
import os

# Download NLTK tokenizer data
nltk.download('punkt')

# Load the Hugging Face model
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline
import torch

# Set your Hugging Face API token
api_token = "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"

# Authenticate with the Hugging Face API
pipeline("text-generation", model="mistralai/Mistral-7B-Instruct-v0.2", revision="main", token=api_token)

# Now you can use the HuggingFacePipeline
model = HuggingFacePipeline.from_model_id(
    model_id="mistralai/Mistral-7B-Instruct-v0.2",
    task="text-generation",
    device=0,  # -1 for CPU, 0 for GPU
    batch_size=2,  # adjust as needed based on GPU map and model size.
    model_kwargs={"temperature": 0, "max_length": 1024, "torch_dtype": torch.bfloat16},
)

# Initialize the Chroma DB and set up the retriever
vectorstore = Chroma(collection_name="summaries", embedding_function=FastEmbedEmbeddings())
store = InMemoryStore()
id_key = "doc_id"
retriever = MultiVectorRetriever(vectorstore=vectorstore, docstore=store, id_key=id_key)

# Define the function to process the PDF and extract text
def process_pdf(uploaded_file):
    # Save the uploaded file to a temporary location
    with tempfile.NamedTemporaryFile(delete=False) as temp_file:
        temp_file.write(uploaded_file.read())
        temp_file_path = temp_file.name

    # Process the PDF
    raw_pdf_elements = partition_pdf(temp_file_path)

    # Remove the temporary file
    os.unlink(temp_file_path)

    text_elements = []
    table_elements = []
    for element in raw_pdf_elements:
        if isinstance(element, str):
            text_elements.append(element)
        elif isinstance(element, list):
            for sub_element in element:
                text_elements.append(sub_element)
        elif isinstance(element, dict) and "table" in element:
            table_elements.append(element["table"])
    return text_elements, table_elements

# Define a function to summarize the text and tables, and store them in Chroma
def summarize_and_store(texts, tables):
    # Combine both text and tables
    elements = texts + tables

    # Prompt template
    prompt_text = """You are an assistant tasked with summarizing tables and text. \
Give a concise summary of the table or text. Table or text chunk: {element} """
    prompt = ChatPromptTemplate.from_template(prompt_text)

    summarize_chain = {"element": lambda x: x} | prompt | model | StrOutputParser()

    # Summarize and store
    element_ids = []
    if elements:
        element_ids = [str(uuid.uuid4()) for _ in elements]
        total_elements = len(elements)
        with st.spinner('Processing document...'):
            for i, s in enumerate(summarize_chain.batch(elements, {"max_concurrency": 2})):
                progress = (i + 1) / total_elements
                st.progress(progress)
                summary_element = Document(page_content=s, metadata={id_key: element_ids[i]})
                retriever.vectorstore.add_document(summary_element)
                retriever.docstore.set(element_ids[i], s)
                st.write(f"Processed element {i+1} of {total_elements}")
    return element_ids

# Define a function to retrieve responses based on user input
def get_response(input_text):
    # Prompt template
    template = """Answer the question based only on the following context, which can include text and tables:
    {context}
    Question: {question}
    """
    prompt = ChatPromptTemplate.from_template(template)
    # RAG pipeline
    chain = (
        {"context": retriever, "question": RunnablePassthrough()}
        | prompt
        | model
        | StrOutputParser()
    )
    response = chain.invoke(input_text)
    return response

# Define the Streamlit app
def main():
    st.title("Document Chat Interface")
    st.write("Upload a PDF and ask questions based on the context of the document.")

    # Upload PDF
    uploaded_file = st.file_uploader("Upload PDF", type="pdf")

    # Process PDF and show text
    if uploaded_file is not None:
        st.write("PDF uploaded successfully!")

        # Process PDF
        text_elements, table_elements = process_pdf(uploaded_file)

        # Display text elements
        st.subheader("Text Elements")
        for text in text_elements:
            st.write(text)

        # Display table elements
        st.subheader("Table Elements")
        for table in table_elements:
            st.write(table)

        # Summarize and store text and tables
        stored_ids = summarize_and_store(text_elements, table_elements)

        # Check if document has been processed and stored
        if stored_ids:
            # Ask question
            questions = st.text_area("Ask questions based on the document context (one question per line):")

            if questions.strip():
                question_list = questions.split('\n')

                st.subheader("Responses:")
                for question in question_list:
                    response = get_response(question)
                    st.write(response)
    else:
        st.info("Please upload a PDF to start.")

if __name__ == "__main__":
    main()
```

```python
import gradio as gr
from unstructured.partition.pdf import partition_pdf
from langchain.retrievers.multi_vector import MultiVectorRetriever
from langchain.storage import InMemoryStore
from langchain_community.vectorstores import Chroma
from langchain_core.documents import Document
from langchain_community.embeddings.fastembed import FastEmbedEmbeddings
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.runnables import RunnablePassthrough
from langchain_community.llms.huggingface_pipeline import HuggingFacePipeline
import torch
import uuid
import nltk
import os
import threading
import logging
import pytesseract
from pydantic import BaseModel
from typing import Any

# Setup logging
logging.basicConfig(level=logging.INFO)

# Download NLTK tokenizer data
nltk.download('punkt')

# Set Tesseract path
pytesseract.pytesseract.tesseract_cmd = (r'/usr/bin/tesseract')

# Initialize the Chroma DB and set up the retriever
vectorstore = Chroma(collection_name="summaries", embedding_function=FastEmbedEmbeddings())
store = InMemoryStore()
id_key = "doc_id"
retriever = MultiVectorRetriever(vectorstore=vectorstore, docstore=store, id_key=id_key)

# PDF file path
pdf_path = "/content/2405.11143v1.pdf"

# Load the Hugging Face model in a separate thread
def load_model():
    global model
    logging.info("Loading Hugging Face model...")
    model = HuggingFacePipeline.from_model_id(
        model_id="mistralai/Mistral-7B-Instruct-v0.2",
        task="text-generation",
        device=0,  # -1 for CPU, 0 for GPU
        batch_size=2,  # adjust as needed based on GPU map and model size.
        model_kwargs={"temperature": 0, "max_length": 1024, "torch_dtype": torch.bfloat16},
    )

# Function to process the PDF and extract text and tables
def process_pdf(pdf_path):
    # Process the PDF
    raw_pdf_elements = partition_pdf(
        filename=pdf_path,
        extract_images_in_pdf=True,
        infer_table_structure=True,
        chunking_strategy="by_title",
        max_characters=4000,
        new_after_n_chars=3800,
        combine_text_under_n_chars=2000
    )
    # Print the extracted elements
    print("Raw PDF Elements:")
    print(raw_pdf_elements)

    category_counts = {}

    for element in raw_pdf_elements:
        category = str(type(element))
        if category in category_counts:
            category_counts[category] += 1
        else:
            category_counts[category] = 1

    # Unique_categories will have unique elements
    unique_categories = set(category_counts.keys())
    category_counts

    class Element(BaseModel):
        type: str
        text: Any

    # Categorize by type
    categorized_elements = []
    for element in raw_pdf_elements:
        if "unstructured.documents.elements.Table" in str(type(element)):
            categorized_elements.append(Element(type="table", text=str(element)))
        elif "unstructured.documents.elements.CompositeElement" in str(type(element)):
            categorized_elements.append(Element(type="text", text=str(element)))

    # Tables
    table_elements = [e for e in categorized_elements if e.type == "table"]
    print(len(table_elements))

    # Text
    text_elements = [e for e in categorized_elements if e.type == "text"]
    print(len(text_elements))

    logging.info("Summarizing and storing text and tables...")
    element_ids = []

    # Prompt template
    prompt_text = """You are an assistant tasked with summarizing tables and text. \
    Give a concise summary of the table or text. Table or text chunk: {element} """
    prompt = ChatPromptTemplate.from_template(prompt_text)

    summarize_chain = {"element": lambda x: x} | prompt | model | StrOutputParser()

    tables = [i.text for i in table_elements]
    table_summaries = summarize_chain.batch(tables, {"max_concurrency": 5})

    # Apply to texts
    texts = [i.text for i in text_elements]
    text_summaries = summarize_chain.batch(texts, {"max_concurrency": 5})

    # Add texts
    doc_ids = [str(uuid.uuid4()) for _ in texts]
    summary_texts = [
        Document(page_content=s, metadata={id_key: doc_ids[i]})
        for i, s in enumerate(text_summaries)
    ]
    retriever.vectorstore.add_documents(summary_texts)
    retriever.docstore.mset(list(zip(doc_ids, texts)))

    # Add tables
    table_ids = [str(uuid.uuid4()) for _ in tables]
    summary_tables = [
        Document(page_content=s, metadata={id_key: table_ids[i]})
        for i, s in enumerate(table_summaries)
    ]
    retriever.vectorstore.add_documents(summary_tables)
    retriever.docstore.mset(list(zip(table_ids, tables)))

    return summary_texts, summary_tables

# Load the model and process the PDF in separate threads
model_thread = threading.Thread(target=load_model)
model_thread.start()

pdf_thread = threading.Thread(target=lambda: process_pdf(pdf_path))
pdf_thread.start()

# Wait for the PDF processing and model loading to complete
pdf_thread.join()
model_thread.join()

# Define the Gradio app
def chat_interface(question):
    # Prompt template
    template = """Answer the question based only on the following context, which can include text and tables:
    {context}
    Question: {question}
    """
    prompt = ChatPromptTemplate.from_template(template)

    # RAG pipeline
    chain = (
        {"context": retriever, "question": RunnablePassthrough()}
        | prompt
        | model
        | StrOutputParser()
    )

    # Now invoke the chain with the specific question
    response = chain.invoke({"context": retriever, "question": question})
    return response

# Launch the Gradio app
gr.Interface(
    fn=chat_interface,
    inputs=["text"],
    outputs=gr.outputs.Textbox(label="Answer"),
    title="Document Chat Interface",
    description="Ask questions based on the content of the provided PDF."
).launch()
```
