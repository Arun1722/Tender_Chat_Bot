```python
from llama_parse import LlamaParse
import glob
import os
import re

# Directory containing your PDFs
pdf_directory = '/content/pdfs/'  # Update this path if your PDFs are in a different directory

# Get all PDF files in the directory
pdf_files = glob.glob(os.path.join(pdf_directory, '*.pdf'))

# Initialize a list to store the parsed content of each document
documents = []

# Parsing instruction
parsing_instruction = """
Extract all text and tabular data from the PDF document and format it in Markdown.

Instructions:
Extract All Text:

Extract the full text content of the PDF, including headings, paragraphs, and other text elements.

Extract All Table Data:

Identify and extract all tables in the PDF, maintaining their original structure.

Output Format:

Use Markdown for the output format.
Maintain clear distinction between different content types (e.g., use different Markdown syntax for headings, paragraphs, tables, etc.).

Markdown Structure:

Document Title and Metadata:
If present, extract the document title and metadata (e.g., bid number, date) and include them at the top of the Markdown document using appropriate Markdown formatting.

Headings and Subheadings:
Use #, ##, ###, etc., for headings and subheadings.

Paragraphs:
Keep paragraphs as plain text separated by a blank line.

Tables:
Format tables using Markdown's table syntax.

Example Usage:
# Extracted PDF Content

**Bid Number:** [Insert Bid Number]
**Date:** [Insert Date]

## Section Heading 1
[Insert paragraph text]

| Header 1  | Header 2  | Header 3  |
|-----------|-----------|-----------|
| Data 1    | Data 2    | Data 3    |
| Data 4    | Data 5    | Data 6    |
| ...       | ...       | ...       |

## Section Heading 2
[Insert paragraph text]

[Repeat for additional content]
"""

# Loop through each PDF file
for pdf_file in pdf_files:
    print(f"Processing {pdf_file}")
    # Parse the PDF with the given instructions
    parsed_documents = LlamaParse(
        result_type="markdown",
        parsing_instruction=parsing_instruction
    ).load_data(pdf_file)

    # Extract text from the Document objects
    if isinstance(parsed_documents, list):
        # Assuming each item is a Document object with a 'text' attribute
        parsed_content = '\n'.join([doc.text for doc in parsed_documents])
    elif isinstance(parsed_documents, str):
        parsed_content = parsed_documents
    else:
        print(f"Unexpected type for parsed_documents: {type(parsed_documents)}")
        parsed_content = ''

    # Store the parsed content in the documents list
    documents.append({
        'filename': os.path.basename(pdf_file),
        'content': parsed_content
    })



for doc in documents:
    print(f"\nDocument: {doc['filename']}")
    print("Content:")
    print(doc['content'])
```

```python
import re

# Function to remove Hindi characters from a text
def remove_hindi_characters(text):
    # Define a regular expression pattern to match Hindi characters
    hindi_pattern = re.compile(r'[\u0900-\u097F]+')
    # Substitute Hindi characters with an empty string
    return hindi_pattern.sub('', text)

# Create a new list to store documents with Hindi characters removed
documents_without_hindi = []

for doc in documents:
    # Remove Hindi characters from the content
    cleaned_content = remove_hindi_characters(doc['content'])
    # Add the cleaned document to the new list
    documents_without_hindi.append({
        'filename': doc['filename'],
        'content': cleaned_content
    })


for doc in documents_without_hindi:
    print(f"\nDocument: {doc['filename']}")
    print("Content:")
    print(doc['content'])
```

```python
import os
from llama_index.core import Document, VectorStoreIndex
from qdrant_client import QdrantClient
from qdrant_client.http.models import VectorParams, Distance, PointStruct
from langchain.embeddings.huggingface import HuggingFaceEmbeddings
from llama_index.vector_stores.qdrant import QdrantVectorStore
from llama_index.core import StorageContext
from langchain.text_splitter import CharacterTextSplitter

# Initialize embedding model
embed_model = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-mpnet-base-v2"
)

# Initialize Qdrant client
client = QdrantClient(
    url="https://fd4ba2d3-e45f-44d6-9b20-b0c5d1cabb9d.us-east4-0.gcp.cloud.qdrant.io",
    api_key="y2EIzplsyAyYJPur5ryVUldfHtT6Q1-rM48c3uVLqsYBhmKk1qPhnQ",
)

# Function to read a Markdown file
def read_markdown(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        return file.read()

# Function to save chunks to separate Markdown files
def save_chunks(chunks, output_dir, base_filename):
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    for i, chunk in enumerate(chunks):
        chunk_filename = f"{base_filename}_chunk_{i+1}.md"
        with open(os.path.join(output_dir, chunk_filename), 'w', encoding='utf-8') as file:
            file.write(chunk)

# Function to chunk text using CharacterTextSplitter
def chunk_text_with_character_splitter(text, chunk_size=512, chunk_overlap=50):
    splitter = CharacterTextSplitter(chunk_size=chunk_size, chunk_overlap=chunk_overlap)
    chunks = splitter.split_text(text)
    return chunks

# Function to store chunks in Qdrant
def store_chunks_in_qdrant(chunks):
    collection_name = "document_GEM_12/08"

    # Check if collection exists and create if necessary
    if not client.get_collection(collection_name):
        client.create_collection(
            collection_name=collection_name,
            vectors_config=VectorParams(size=768, distance=Distance.COSINE)
        )

    for doc_id, chunk in enumerate(chunks):
        embedding = embed_model.embed_query(chunk)

        client.upsert(
            collection_name=collection_name,
            points=[
                PointStruct(
                    id=doc_id,
                    vector=embedding,  # Pass the list directly
                    payload={"text": chunk}
                )
            ]
        )

# Main execution
if __name__ == "__main__":
    input_file = "/content/NEW_11_prompt.md"
    output_directory = "/content/chunks"

    markdown_text = read_markdown(input_file)
    chunks = chunk_text_with_character_splitter(markdown_text)
    save_chunks(chunks, output_directory, os.path.splitext(os.path.basename(input_file))[0])
    store_chunks_in_qdrant(chunks)

    print(f"Saved {len(chunks)} chunks to '{output_directory}' directory and stored them in Qdrant.")
```

```python
from flask import Flask, request, jsonify
import time


app = Flask(__name__)


@app.route('/api/v1/qa/', methods=['POST'])
def query_engine_api():
    data = request.get_json()
    query = data.get('query', '')
    start_time = time.time()
    response = query_engine.query(query)
    elapsed_time = round(time.time() - start_time, 2)
    response_data = {
        "response": str(response),
        "elapsed_time": f"{elapsed_time}s"
    }

    return jsonify(response_data)


if __name__ == '__main__':
    app.run(host='0.0.0.0', port=5000)
```
